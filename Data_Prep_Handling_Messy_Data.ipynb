```python
import pandas as pd
import numpy as np

# Create DataFrame
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

# ---- Impute missing values ----
# Age → median
df['age'].fillna(df['age'].median(), inplace=True)

# City → mode
df['city'].fillna(df['city'].mode()[0], inplace=True)

# ---- Remove duplicates ----
df = df.drop_duplicates()

print("Cleaned Data:")
print(df)
```

Task 2 — Encode Categorical Columns

```python
from sklearn.preprocessing import LabelEncoder

# ---- One-Hot Encoding for 'city' ----
df = pd.get_dummies(df, columns=['city'])

# ---- Label Encoding for 'gender' ----
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

print("\nEncoded Data:")
print(df)
```

Task 3 — Scale Numerical Features

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Copy dataset for both methods
df_minmax = df.copy()
df_standard = df.copy()

# Columns to scale
cols = ['age', 'annual_income']

# ---- Min-Max Scaling ----
minmax = MinMaxScaler()
df_minmax[cols] = minmax.fit_transform(df_minmax[cols])

print("\nMin-Max Scaled Data:")
print(df_minmax)

# ---- Standard Scaling ----
standard = StandardScaler()
df_standard[cols] = standard.fit_transform(df_standard[cols])

print("\nStandard Scaled Data:")
print(df_standard)
```

When to Use Which Scaling?
Min-Max Scaling is preferred when you need values in a fixed range (0–1), especially for models like neural networks or distance-based algorithms (KNN).
StandardScaler (Z-score) is better when data has outliers or unknown distribution, and is commonly used in models like Logistic Regression or SVM.
